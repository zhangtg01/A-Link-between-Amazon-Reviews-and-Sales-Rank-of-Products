# Review Data Processing & Polarity Scores Worksheet

# Import all necessary libraries

```python
import pandas as pd
import nltk
nltk.download('punkt')
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.tokenize import word_tokenize
from textblob import TextBlob
```

## Importing the data

```python
df = pd.read_csv ('review_data_cleaned.csv')
df.head(-10)
```

## Remove stopwords and get the polarity scores based on sentiment analysis

```python
df_reviewText = df[['asin','reviewText']]
df_reviewText.head()
```

```python
#Tokenize & remove stopwords from review text row entries
final_list = []
#Borrow the code from https://www.geeksforgeeks.org/removing-stop-words-nltk-python/
stop_words = set(stopwords.words('english'))

for index, item in df_reviewText['reviewText'].iteritems():
    #Borrow the code from https://www.geeksforgeeks.org/removing-stop-words-nltk-python/
    middle = []
    middle.append(word_tokenize(str(item)))
    for i in middle:
        filtered_list = []
        for j in i:
            if (j.lower() not in stop_words) and (j.isalnum() and len(j) > 1) == True:
                filtered_list.append(j)
            else:
                continue
    final_list.append(filtered_list)
```

```python
len(final_list)
```

```python
y = pd.Series(final_list)
y.head()
```

```python
#df_reviewText = df_reviewText.drop(columns='ready_for_sentiment_analysis')
```

```python
df_reviewText.insert(0, 'summary_preprocessed', y)
```

```python
df_reviewText.head()
```

```python
# Calculate polarity scores on review text for each row in dataframe
pol_list = []
for index, item in df_reviewText['ready_for_sentiment_analysis'].iteritems():
    #Borrow the code from nlp_part3_sentiment_analysis-inclass.ipynb
    if item == '':
        continue
    else:
        tb = TextBlob(str(item))
        pol_list.append(tb.sentiment.polarity)
```

```python
df['Review_polarity']=pol_list
df.head()
```

```python
df_summary = df[['summary']]
df_summary.head()
```

```python
#Tokenize & remove stopwords from Summary text row entries
final_list_sm = []
#Borrow the code from https://www.geeksforgeeks.org/removing-stop-words-nltk-python/
stop_words = set(stopwords.words('english'))

for index, item in df_summary['summary'].iteritems():
    #Borrow the code from https://www.geeksforgeeks.org/removing-stop-words-nltk-python/
    middle_sm = []
    middle_sm.append(word_tokenize(str(item)))
    for i in middle_sm:
        filtered_list_sm = []
        for j in i:
            if (j.lower() not in stop_words) and (j.isalnum() and len(j) > 1) == True:
                filtered_list_sm.append(j)
            else:
                continue
    final_list_sm.append(filtered_list_sm)
```

```python
len(final_list_sm)
final_list_sm[:5]
```

```python
y_sm = pd.Series(final_list_sm)
y_sm.head()
```

```python
df_summary.insert(0, 'ready_for_sentiment_analysis_sm', y_sm)
```

```python
df_summary.head()
```

```python
# Calculate polarity scores on summary text for each row in dataframe
pol_list_sm = []

for index, item in df_summary['ready_for_sentiment_analysis_sm'].iteritems():
    #Borrow the code from nlp_part3_sentiment_analysis-inclass.ipynb
    if item == '':
        continue
    else:
        tb = TextBlob(str(item))
        pol_list_sm.append(tb.sentiment.polarity)
```

```python
z_sm = pd.Series(pol_list_sm)
z_sm.head()
```

```python
df['summary_polarity']=z_sm
```
